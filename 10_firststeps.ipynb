```python
import pandas as pd
import altair as alt

alt.renderers.enable("mimetype")
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/onlyphantom/miband/main/data/run_1km.csv", parse_dates=['startTime'])
df['day_of_week'] = df['startTime'].dt.day_name()
df['week'] = df['startTime'].dt.week
```

```python
# Basis Scatter plot
alt.Chart(df).mark_point().encode(
    x = "seconds_per_km",
    y = "day_of_week",
    color = "day_of_week"
)
```

```python
# interaktiv
alt.Chart(df).mark_point().encode(
    x = "seconds_per_km",
    y = "day_of_week",
    color = "day_of_week"
).interactive()
```

```python
# Statt Keywords - mit Konstruktoren arbeiten um mehr Optionen zu erhalten
alt.Chart(df).mark_point().encode(
    alt.X("seconds_per_km:Q"),
    alt.Y("day_of_week:N")) #Nicht ntig, aber zum erinnnern: mit :Type kann man sagen was das fr eine Variable ist; Q;N;O; T
```

```python
# Einfache Aggregation
alt.Chart(df).mark_point().encode(
    x = "average(seconds_per_km)", # es ginge auch min, max, count median stdev
    y = "day_of_week",
    color = "day_of_week"
)
```

```python
# "Geoms" bzw. Marker design oder Achsen
alt.Chart(df).mark_point(color="chartreuse").encode( # jede CCS Color oder HEX Code
    alt.X("stdev(seconds_per_km)", title="BlaBlaBla", scale=alt.Scale(type="log")), # ginge auvh noch tiefer mit alt.Axis siehe bei Y-Achse
    alt.Y("day_of_week", axis=alt.Axis(title="Wochentag"))
)
```

```python
alt.Chart(df).mark_line(color="lime").encode(
    alt.X("week"),
    alt.Y("average(seconds_per_km)", title="Speed")
    
)
```

```python
# mit punkten (umstndlich aber didaktisch wertvoll :-))
line = alt.Chart(df).mark_line(color="aqua").encode(
    alt.X("week"),
    alt.Y("average(seconds_per_km)", title="Speed")
    
)
points = alt.Chart(df).mark_circle(color="magenta").encode(
    alt.X("week"),
    alt.Y("average(seconds_per_km)", title="Speed")
    
)
line + points
```

```python
# mit punkten BESSER
line = alt.Chart(df).mark_line(color="aqua").encode(
    alt.X("week"),
    alt.Y("average(seconds_per_km)", title="Speed")
    
)
line + line.mark_circle(color="magenta")
```

```python
# Da --o-- sehr gebruchlich ist gbe es auch:
alt.Chart(df).mark_line(color="magenta", point=True,).encode(
    alt.X("week"),
    alt.Y("average(seconds_per_km)", title="Speed")
    
)
```

```python
## Layout
speed = alt.Chart(df).mark_line(color="aqua", point=True).encode(
    alt.X("week"),
    alt.Y("average(seconds_per_km)", title="Speed")
    
)

dist = speed.encode(  # erbt alles von speed bis auch Y Achse!
    alt.Y("sum(distance(m))", title="Distanz")
)

# Concat "&" wrde unerhalb sein
speed | dist
```

```python
# mehr zur interaktion

speed = alt.Chart(df).mark_point(color="aqua").encode(
    alt.X("week"),
    alt.Y("average(seconds_per_km)", title="Speed"),
    tooltip=["date", "distance(m)"]
    
)

speed
```

```python
alt.Chart(df).mark_bar().encode(
    alt.X("count(speed_per_km)"),
    alt.Y("month(startTime):O") ## monat kann altair selbst ermitteln, wenn richtiger datatype / das O sortiert dann noch
)
```

```python
brush = alt.selection_interval(encodings=["y"])
auswahl_farbe = alt.condition(brush, alt.value(1), alt.value(0.4))

runs = alt.Chart(df).mark_bar(color="aqua").encode(
    alt.X("count(speed_per_km)"),
    alt.Y("month(startTime):O", title=None, timeUnit="month"), # timeUnit nicht ntig hier
    tooltip=["month(startTime)", "count(speed_per_km)"],
    opacity = auswahl_farbe

).add_selection(brush)

runs
```

```python
# Baut auf "runs" von oben auf
brush = alt.selection_interval(encodings=["y"])
auswahl_farbe = alt.condition(brush, alt.value(1), alt.value(0.4))

runs = alt.Chart(df).mark_bar(color="aqua").encode(
    alt.X("count(speed_per_km)"),
    alt.Y("month(startTime):O", title=None, timeUnit="month"), # timeUnit nicht ntig hier
    tooltip=["month(startTime)", "count(speed_per_km)"],
    opacity = auswahl_farbe
).add_selection(brush).properties(
    width=300,
    height=250
)

speed = alt.Chart(df).mark_circle(color="magenta").encode(
    alt.X("week", title="Woche"),
    alt.Y("seconds_per_km"),
    opacity = auswahl_farbe
).properties(
    width=300,
    height=250
)


runs | speed
```
